```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_val_score
```

```python
m_data = pd.read_csv('/content/melb_data.csv')
```

```python
m_data.describe()
```

```python
m_data.columns
```

```python
m_data.shape
```

```python
m_data.notnull()
```

```python
m_data=m_data.dropna(axis=0)
```

```python
m_data.isnull()
```

```python
m_data.shape
```

```python
y = m_data.Price
```

```python
y
```

```python
m_data.columns
```

```python
features = ['Rooms','Bedroom2','Bathroom','Landsize','Lattitude','Longtitude']
```

```python
x = m_data[features]
```

```python
x
```

```python
x.describe()
```

```python
model = LinearRegression()
```

```python
train_x, test_x, train_y, test_y = train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
scaler = StandardScaler()
scaler_x = scaler.fit_transform(train_x)
scaled_y = scaler.fit_transform(test_x)
```

```python
model.fit(scaler_x,train_y)
```

```python
p_x = model.predict(scaled_y)
```

```python
mae_mae = mean_absolute_error(p_x,test_y)
```
